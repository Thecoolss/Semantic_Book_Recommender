```python
import pandas as pd
import numpy as np
books = pd.read_csv("books_cleaned.csv")
```

```python
books["categories"].value_counts().reset_index().query("count > 50")
category_mapping = {'Fiction' : "Fiction",
 'Juvenile Fiction': "Children's Fiction",
 'Biography & Autobiography': "Nonfiction",
 'History': "Nonfiction",
 'Literary Criticism': "Nonfiction",
 'Philosophy': "Nonfiction",
 'Religion': "Nonfiction",
 'Comics & Graphic Novels': "Fiction",
 'Drama': "Fiction",
 'Juvenile Nonfiction': "Children's Nonfiction",
 'Science': "Nonfiction",
 'Poetry': "Fiction",
 "Business":"Nonfiction",
 "Literacy Collections":"Nonfiction",}

books["simple_categories"] = books["categories"].map(category_mapping)
```

```python
from transformers import pipeline
```

```python
simple_cats=["Fiction","Nonfiction"]
classifier=pipeline("zero-shot-classification",
                    model="facebook/bart-large-mnli")
                    
```

```python
import numpy as np
def get_prediction(sequence,categories):
  prediction=classifier(sequence,categories)
  return prediction["labels"][np.argmax(prediction["scores"])]

sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[0]
get_prediction(sequence,simple_cats)
```

```python
actual_cats=[]
pred_cats=[]

for i in range(200):
  sequence=books.loc[books["simple_categories"]=="Fiction","description"].reset_index(drop=True)[i]
  pred_cats+=[get_prediction(sequence,simple_cats)]
  actual_cats+=["Fiction"]

pred_cats
```
